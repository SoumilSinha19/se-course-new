```python
import matplotlib.pyplot as plt
import numpy as np

def quadratic_model(time, a, b, c):
 temperature = a * time**2 + b * time + c
 return temperature

# Get coefficients from user input
a_input = float(input("Enter coefficient a: "))
b_input = float(input("Enter coefficient b: "))
c_input = float(input("Enter coefficient c: "))

time_values = np.linspace(0, 10, 100)
temperature_input = quadratic_model(time_values, a_input, b_input,
c_input)

# Plotting
plt.plot(time_values, temperature_input, label='User input coefficients')

# Plot settings
plt.xlabel('Time')
plt.ylabel('Temperature')
plt.legend()
plt.title('Weather Modeling with Quadratic Equation (Single Input)')
plt.show()
```
